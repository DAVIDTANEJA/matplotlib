# Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['Google','yahoo','microsoft','ferrari']
revenue=[125,100,90,70]
profit=[40,20,34,12]
```

```python
xpos=np.arange(len(company))
plt.xticks(xpos,company)           # here xpos=0,1,2,3 changed by-> company names

plt.ylabel('Revenue($)')
plt.title('US Tech. Stocks')

plt.bar(xpos-0.2,revenue,label='Revenue', width=0.4)    # color='r'
plt.bar(xpos+0.2,profit,label='Profit', width=0.4)   
# it shows revenue and profit in same graph , bcoz x-position of company is same, 
# now we use 'xpos' array -0.2 and +0.2 to separate revenue and profit and width func. also


plt.legend()

plt.show()
```

## horizontal bar chart , so function is ->  barh() 

```python
xpos=np.arange(len(company))
plt.yticks(xpos,company)     # here change yticks ,so company names come
plt.xlabel('Revenue($)')
plt.title('US Tech Stocks')

plt.barh(xpos-0.2,revenue,label='Revenue')      # barh()
plt.barh(xpos+0.2,profit,label='Profit')

plt.legend()
plt.show()
```

#  Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# histogram- need only single dimension array and  bins(or bucket) - is nothing but a range

blood_sugar_man=[113,85,90,150,149,88,93,115,135,80,77,82,129]   
blood_sugar_woman=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80,100,125,200], rwidth=0.95, color=['g','orange'],
        label=['man','woman'],orientation='horizontal')                                                    # histtype
plt.legend()
# it creates 10 bins, we can change it bins=3 , also we can specify ranges 
# like:80-100 normal , 100-125 pre-diabetic , 125 above diabetic
# rwidth(0 to 1)- relative width compare to bin size , to separate bars 
# histtype - shows step  (histtype='step')

# orientation='horizontal' , place the histogram horizontally  

plt.xlabel('sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood sugar Analysis')
```

```python
# create histogram plots and figures using --->  Numpy and Matplotlib

from IPython import get_ipython                           # magic code
get_ipython().run_line_magic('matplotlib', 'inline')

from numpy import random
import numpy as np
import matplotlib.pyplot as plt

# mu - assigned to mean of statistical distribution
# sigma - assigned to standard deviation
mu, sigma = 100, 15        # mean - 100  , std - 15
data_set = mu + sigma * random.randn(10000)    # using 10000 data points from data set            

# matplotlib - histogram function
n, bins, patches = plt.hist(data_set, 50, normed=1, facecolor='g', alpha=0.75)
# bins - 50 , value of datapoints will be normalised to 1 , g-green , tansparency set -0.75

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### stack plots

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]      # x-axis

sleeping = [7,8,5,6,7,9,8]       # these all 4 goes to y-axis
eating = [1,3,4,2,5,3,2]
working = [9,11,7,11,5,10,11]
playing = [2,1,4,3,5,2,4]


plt.plot([],[], color='r', label='Sleeping')         # these 4 using for legend to know which pat belongs to
plt.plot([],[], color='b', label='Eating')
plt.plot([],[], color='c', label='Working')
plt.plot([],[], color='m', label='Playing')


plt.stackplot(days, sleeping,eating,working,playing, colors=['r','b','c','m'])   
# days= x-axis, all 4 are y-axis, colors for y-axis 4 color

plt.legend(loc='best', bbox_to_anchor=(1,0.5,0.5,0.5))     # bbox_to_anchor=()  - using to put legends outside
plt.show()
```

# Pie chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_values=[1400,600,300,410,250]
category=['Home-Rent','Food','Phone-Bill','Car','Other-Utilities']
```

```python
plt.axis("equal")   # data units for x-y are same , but in pixels are different

plt.pie(exp_values, labels=category, radius=2, autopct='%0.0f%%', shadow=True, explode=[0,0.2,0,0,0.3], startangle=180)  
# labels = category -used to add labels ,
# radius-used to increase size,
# autopct=%f%% -adds percentage in chart ,(%0.0f%%)-gives only percentage value,(0.1)- decimal 1 value,(0.2)-2 decimal values
# explode - takes values = labels , when we change value of any like: 0 to 0.2 (of food) it comes out in chart 
# startangle=45 , changes angle of starting Like: home-rent start from 45 degree

plt.show()
```

## Plot Annotations

##### An annotation is extra information associated with a particular point in a document or other piece of information. It can be a note that includes a comment or explanation

```python
import matplotlib.pyplot as plt
import numpy as np

number_of_data_points = 10

my_figure = plt.figure()
subplot_1 = my_figure.add_subplot(1,1,1)
subplot_1.plot(np.random.rand(number_of_data_points).cumsum())

subplot_1.text(1, 0.5, r'an equation: $E=mc^2$', fontsize=18, color='red')
subplot_1.text(1, 1.5, "Hello, Mountain climbing!", family='monospace', fontsize=18, color='green')

# see : http://matplotlib.org/users/transforms_tutorial.html
# transform=subplot_1.transAxes;entire axis between zero and one

subplot_1.text(0.5, 0.5, "we are centered,now", transform=subplot_1.transAxes)

subplot_1.annotate("show arrow", xy=(2,1), xytext=(3,4),
                  arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```

## live graphs 
#### add data to file it updated on graph

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style

# print(style.available)
plt.style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('example.txt', 'r').read()
    lines = graph_data.split('\n')
    xs=[]
    ys=[]
    zs=[]
    for line in lines:
        if len(line)>1:
            x,y,z = line.split(',')
            xs.append(x)
            ys.append(y)
            zs.append(z)
    ax1.clear()
    ax1.plot(xs,ys,zs)
    
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```

### subplots ->  fig.add_subplot()  ,  plt.subplot2grid()

```python
import random
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use('fivethirtyeight')

fig = plt.figure()      # figure 

xs=[]                   
ys=[]                          # data
for i in range(10):
    x = random.randrange(10)
    y = i
    xs.append(x)
    ys.append(y)

ax1 = fig.add_subplot(221)      # (height, width, no. of plot) 
ax2 = fig.add_subplot(222) 
ax3 = fig.add_subplot(212) 

ax1.plot(xs,ys, 'r-')
ax2.plot(xs,ys, 'b--')
ax3.plot(xs,ys)

plt.show()
```

```python
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=2, colspan=1)                # plt.subplot2grid()
ax2 = plt.subplot2grid((6,1), (2,0), rowspan=2, colspan=1)
ax3 = plt.subplot2grid((6,1), (4,0), rowspan=6, colspan=1)

ax1.plot(xs,ys)
ax2.plot(xs,ys)
ax3.plot(xs,ys)
plt.show()
```
